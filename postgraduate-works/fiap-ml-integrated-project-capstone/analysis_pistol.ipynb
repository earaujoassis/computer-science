```python
import matplotlib
import numpy as np
import pandas as pd
import matplotlib.pyplot as plt

%matplotlib inline
```

```python
import os
import json


def get_events_league_json(mapname, leaguename):
    pathname = '{0}/Events'.format(mapname)

    files = os.listdir(pathname)
    objects_str = []
    for filename in files:
        if not filename.startswith(leaguename):
            continue
        filepath = '{0}/{1}'.format(pathname, filename)
        with open(filepath) as f:
            accumulator = ''
            for line in f:
                if line.strip() == '}':
                    accumulator += line.strip()
                    objects_str.append('{0}'.format(accumulator))
                    accumulator = ''
                else:
                    accumulator += line.strip()

    return objects_str


def from_text_to_json_object(text):
    return json.loads(text)
```

```python
objects = get_events_league_json('de_mirage', 'pro_events')
#objects = get_events_league_json('de_mirage', 'alfa_events')
#objects = get_events_league_json('de_mirage', 'amadora_events')
```

# Análise de Round Pistol

## Mirage
### Alfa
- maior número de rounds sendo decididos por objetivo, com as bombas sendo desarmadas na maioria das vezes;
- normalmente um time em específico venceu ambos os lados do pistol;

### Pros
- equilibrado o número de rounds em que as bombas explodiram e foram desarmadas;
- normalmente o pistol favorece o time TR;

## Cache
### Alfa
- maior número de rounds vencidos por kill, sendo o lado TR que mais prevalece;
- normalmente um time em específico venceu ambos os lados do pistol ou o lado TR prevaleceu;

### Pros
- maior número de rounds decididos por explosão de bomba (TR prevalecendo novamente);
- normalmente o pistol favorece o time TR.
